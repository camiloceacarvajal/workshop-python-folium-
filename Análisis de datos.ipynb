### Esta lección se enfocará en la limpieza de datos y visualizacion de los datos.

- CSV 
significa valores separados por comas. Estos son los archivos más comunes y ampliamente utilizados con un DataFrameobjeto pandas . Es un tipo de archivo de texto que usa la extensión .csv en lugar de .txt . El texto está separado por comas , en este archivo, y cuando un DatFramelee este archivo, los valores entre comas se convierten en valores de celda de DataFrame.

- Ṕandas paquete proporciona funciones de lectura y escritura para archivos csv .

**Lectura de archivos CSV**

- La read_csv(file_name)función se utiliza para leer desde un archivo csv . El file_namedebe ser reemplazado con el nombre de su archivo. Asegúrese de que el archivo está en el mismo directorio que el archivo de programa o claramente definir la ruta a ese archivo, por ejemplo, folder/folder/file.csv. Si no se define una ruta con el nombre del archivo, el programa busca el archivo en el directorio actual.

<h2>Leer / guardar otros formatos de datos</h2>



| Formato Dato  | Leer           | Guardar            |
| ------------- |:--------------:| ----------------:|
| csv           | `pd.read_csv()`  |`df.to_csv()`     |
| json          | `pd.read_json()` |`df.to_json()`    |
| excel         | `pd.read_excel()`|`df.to_excel()`   |
| hdf           | `pd.read_hdf()`  |`df.to_hdf()`     |
| sql           | `pd.read_sql()`  |`df.to_sql()`     |

```python
import folium
from folium.plugins import MarkerCluster
from folium import plugins
from folium.plugins import MeasureControl
import branca
import numpy as np
import vincent
import os
import json
from folium.plugins import HeatMap
```

```python
attr = ('&copy; <a href="http://www.openstreetmap.org/copyright">OpenStreetMap</a> '
        '&copy; <a href="http://www.cr2.cl/datos-de-caudales/?cp_Caudal=2/copyright">Resilencia_CR2</a>'
        '&copy; <a href="https://n9.cl/lw81/attributions">CamiloGithub</a>')
m = folium.Map(resu_cr2[['latitud','longitud']].mean().tolist(),
               attr=None,tiles = None, zoom_start=6,control_scale=True,prefer_canvas=True,detect_retina=True)
folium.raster_layers.TileLayer(
    tiles='http://{s}.google.com/vt/lyrs=s&x={x}&y={y}&z={z}',
    attr= attr,name='Google-Satelite',
    max_zoom=20,subdomains=['mt0', 'mt1', 'mt2', 'mt3'],
    overlay=False,control=True,
).add_to(m)
folium.raster_layers.TileLayer(
    tiles='https://a.tile.opentopomap.org/{z}/{x}/{y}.png',
    attr= attr,name='OSM-TOPO',
    max_zoom=20,subdomains=['mt0', 'mt1', 'mt2', 'mt3'],
    overlay=False,control=True,
).add_to(m)
plugins.Fullscreen(
    position='topright',title='EXPANDIR',
    title_cancel='SALIR',force_separate_button=True).add_to(m)
mc = MarkerCluster(name="Ubicacion Estacion ")
for N,data in resu_cr2.iterrows():
    station_html = folium.Html('<b>%s</b>' %(pd.DataFrame(data).to_html()),script=True)
    mc.add_child(folium.Marker([data.latitud,data.longitud],
 popup = folium.Popup(station_html),icon=folium.Icon(color='green'if data.codigo_cuenca  == '83'
                  else 'red',prefix = 'fa',icon='area-chart')                              
                    )) 
df_copy = resu_cr2.copy()
HeatMap(data=df_copy[['latitud', 'longitud']].groupby(['latitud', 'longitud']).sum().reset_index().values.tolist(), radius=11, max_zoom=13,name='Concentracion De Estaciones').add_to(m)
m.add_child(mc)
m.add_child(folium.LayerControl())
m.save(os.path.join( 'Proyectoclase_caudal.html'))
m
```
